```python
class NoeudJeu:
    def __init__(self, joueur, information):
        self.joueur = joueur
        self.information = information
        self.enfants = []

    def ajouter_enfant(self, action, information_enfant):
        enfant = NoeudJeu(joueur=self.joueur % 2 + 1, information=information_enfant)
        self.enfants.append((action, enfant))

def construire_arbre_jeu():
    racine = NoeudJeu(joueur=1, information="Début du jeu")

    for carte_joueur1 in range(1, 6):
        noeud_joueur1 = NoeudJeu(joueur=1, information=f"Joueur 1 joue {carte_joueur1}")

        for carte_joueur2 in range(1, 6):
            noeud_joueur2 = NoeudJeu(joueur=2, information=f"Joueur 2 joue {carte_joueur2}")

            resultat = determine_gagnant(carte_joueur1, carte_joueur2)
            information_enfant = f"Résultat : {resultat}"

            noeud_joueur1.ajouter_enfant(carte_joueur2, information_enfant)
            noeud_joueur2.ajouter_enfant(carte_joueur1, information_enfant)

        racine.enfants.append((carte_joueur1, noeud_joueur1))

    return racine

def determine_gagnant(carte_joueur1, carte_joueur2):
    if carte_joueur1 > carte_joueur2:
        return "Joueur 1 gagne !"
    elif carte_joueur1 < carte_joueur2:
        return "Joueur 2 gagne !"
    else:
        return "Match nul."

def afficher_arbre(noeud, indentation=0):
    print("  " * indentation, f"Joueur {noeud.joueur}: {noeud.information}")

    for action, enfant in noeud.enfants:
        print("  " * (indentation + 1), f"Action {action}:", end=" ")
        afficher_arbre(enfant, indentation + 2)

# Construction de l'arbre du jeu
arbre_jeu = construire_arbre_jeu()

# Affichage de l'arbre du jeu
afficher_arbre(arbre_jeu)
```
